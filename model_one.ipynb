```python
%pylab inline
import re
import math
import string
import collections
from collections import Counter
from __future__ import division
import csv
```

```python
def words(text):
    result = re.findall('[a-z]+', text.lower())
    return result
```

```python
def known(words):
    #Return the subset of words that are actually in the dictionary."
    return {w for w in words if w in COUNTS}

def edits0(word): 
    return {word}

def edits2(word):
    return {e2 for e1 in edits1(word) for e2 in edits1(e1)}
```

```python
def edits1(word):
    pairs      = splits(word)
    deletes    = [a+b[1:]           for (a, b) in pairs if b]
    transposes = [a+b[1]+b[0]+b[2:] for (a, b) in pairs if len(b) > 1]
    replaces   = [a+c+b[1:]         for (a, b) in pairs for c in alphabet if b]
    inserts    = [a+c+b             for (a, b) in pairs for c in alphabet]
    return set(deletes + transposes + replaces + inserts)

def splits(word):
    return [(word[:i], word[i:]) 
            for i in range(len(word)+1)]

alphabet = 'abcdefghijklmnopqrstuvwxyz'
```

```python
def correct(word):
    #Generating all the words with edit distance of 0, 1 & 2
    candidates = (known(edits0(word)) or 
                  known(edits1(word)) or 
                  known(edits2(word)) or 
                  [word])
    return max(candidates, key=COUNTS.get)
```

```python
def correct_text(text):
    return re.sub('[a-zA-Z]+', correct_match, text)

def correct_match(match):
    word = match.group()
    return case_of(word)(correct(word.lower()))

def case_of(text):
    return (str.upper if text.isupper() else
            str.lower if text.islower() else
            str.title if text.istitle() else
            str)
```

```python
# big text
source1 = open('big.txt').read()
word1 = words(source1)
```

```python
# spell-errors.txt
source2 = open('spell-errors.txt').read().split('\n')
word2 = []
for lineString in source2:
    temp = lineString.split(':')
    word2.append(temp[0])
```

```python
# count_1w.txt
source3 = open('count_1w.txt').read().split('\n')
word3 = []
for lineString in source3:
    temp = lineString.split('\t')
    word3.append(temp[0])
```

```python
# count_2w.txt
source4 = open('count_2w.txt').read().split('\n')
word4 = []
for line in source4:
    line = line.split('\t')
    for token in line:
        word4.append(token)

word4
```

```python
wordBank = word1 + word2 + word3 + word4

wordBank
```

```python
COUNTS = Counter(wordBank)
```

```python
testCSV = open('test.csv' , 'r')
csvCursor = csv.reader(testCSV)

output = []
for row in csvCursor:
    output.append(row)

testCSV.close()
```
